```python
#feature selection
import seaborn as sns
import matplotlib.pyplot as plt
#get correlations of each features in dataset
corrmat = data.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20,20))
#plot heat map
g=sns.heatmap(data[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```

```python
#get correlations of each features in dataset
corrmat = data.corr()
top_corr_features = corrmat.index

#filter out features with correlation less than 0.2
high_corr_features = corrmat[abs(corrmat) > 0.2].stack().reset_index().rename(columns={0:'correlation'}).dropna(subset=['correlation'])

#select only the unique high-correlation features
high_corr_features = high_corr_features[high_corr_features['level_0'] != high_corr_features['level_1']].drop_duplicates(subset='correlation', keep='first')

#extract the feature names
selected_features = high_corr_features['level_0'].unique().tolist() + high_corr_features['level_1'].unique().tolist()

print("Selected Features: ", selected_features)
```

```python
# Select only the selected features from the main dataset
selected_data = data[selected_features]

# Print the first few rows of the selected data to verify that it worked
print(selected_data.head())
```

It is generally a good practice to create the correlation matrix between all the features (both independent and dependent variables) before separating the target variable and the independent variables. This is because the correlation between the independent variables may have an impact on the target variable, so it is important to consider these correlations when selecting the features.

Once you have created the correlation matrix, you can use it to select the most meaningful features for your model. This can be done using various techniques, such as selecting the features with the highest correlation to the target variable, or selecting features with the highest intercorrelations.

After selecting the meaningful features, you can then separate the target variable and the independent variables and use the selected features to build your model.
